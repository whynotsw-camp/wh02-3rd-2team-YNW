```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import time

# 웹 드라이버 설정
driver = webdriver.Chrome()
actions = ActionChains(driver)

category = "ORDER"
category = "CANCEL"

driver.get(f"https://mc.coupang.com/ssr/desktop/contact/faq?categoryCode={category}")

time.sleep(3)
print("사이트 출력됨")

# "더보기" 버튼 모두 클릭
try:
    while True:
        plus = driver.find_element(By.CSS_SELECTOR, "div.sc-yvmyzy-0.gGJNMm")
        actions.move_to_element(plus).click().perform()
        print("더보기 클릭됨")
        time.sleep(2)
except:
    print("더보기 모두 클릭함")

# 질문 및 답변 데이터를 저장할 리스트 초기화
question_list = []
answer_list = []

# 모든 질문 요소 가져오기
questions = driver.find_elements(By.CSS_SELECTOR, "li.sc-w7jy57-0.kLBJZv")

# 질문 및 답변 크롤링

main_q = ""
main_ans = ""
sub_q = []
sub_ans = [] 

for elem in questions:
    try:
        main_q = ""
        main_ans = ""
        sub_q = []
        sub_ans = [] 

        # 질문 텍스트 추출
        main_q = elem.text.strip()
        print("질문:", main_q)  # 확인용 출력
        # question_list.append(title)

        # Q에 대한 A 출력 버튼 클릭
        button = elem.find_element(By.CSS_SELECTOR, "span.sc-1qu6ltb-3.eobgeA")
        actions.move_to_element(button).click().perform()
        time.sleep(5)

        # 답변 크롤링

        main_ans = elem.find_element(By.CSS_SELECTOR, "div.sc-to1spv-3.kKzMQa.ck-content").text.strip()
        main_ans = main_ans.replace("&nbsp;", "")
        print("답변:", main_ans)  # 확인용 출력
        # answer_list.append(main_ans)

        data = elem.find_elements(By.TAG_NAME, "details") # 얘는 길이가 0인 리스트로 나옴 

        if len(data) == 0:
            print("세부 클릭 정보 없음")
            question_list.append(main_q)
            answer_list.append(main_ans)
            continue

        for datum in data:
            print("sub question: ", datum.text)
            sub_q.append(datum.text)
            
            # 디테일 데이터 
            actions.move_to_element(datum).click().perform()
            time.sleep(5)
            print("디테일 데이터 클릭함")

            # 디테일 데이터가 뜨면, 디테일 크롤링 
            detail_text = datum.find_element(By.CSS_SELECTOR, "div.detail-box-description").text
            print("sub_answer: ", detail_text)
            sub_ans.append(detail_text)


        for question in sub_q:
            question_list.append(main_q + " + " + question)
        for ans in sub_ans:
            answer_list.append(main_ans + " + " + ans)
            
    except Exception as e:
        print("에러 발생:", e)
        answer_list.append("답변 없음")  # 답변이 없는 경우 처리

# 브라우저 종료
driver.quit()

# 데이터프레임 생성 및 저장
df = pd.DataFrame({"Question": question_list, "Answer": answer_list})
df.to_csv(f"{category}_main_sub_qa_list.csv", index=False, encoding="utf-8-sig")

print(f"CSV 파일 저장 완료: {category}_main_sub_qa_list.csv")
```
